Library import

```python
import pandas as pd
import numpy as np

import re
from datetime import datetime, timedelta

import warnings    
warnings.filterwarnings("ignore") 
```

Reading .csv file from resort_stage1.ipynb

```python
resort_df = pd.read_csv("../Data/resort_stage1.csv")
```

```python
resort_df.head()
```

```python
print(resort_df.iloc[0].apply(type))
```

### Location: "List" format to String (removing brackets)

```python
resort_df = resort_df.applymap(lambda x: x[2 : -2] if isinstance(x, str) and x.startswith("[") and x.endswith("]") else x)
```

### Location: get City and Canton

```python
resort_df["City"] = resort_df.Location.str.split("–", expand = True)[0]
resort_df["Canton"] = resort_df.Location.str.split("–", expand = True)[1]
```

```python
resort_df.head()
```

### URL: incorrect format 

```python
resort_df["Link"][0]
```

```python
resort_df["Link"] = resort_df["Link"].str.replace("//", "/") 
resort_df["Link"] = resort_df["Link"].str.replace("https:/", "https://")   
```

```python
resort_df["Link"][0]
```

### String split: remove unnecessary info

```python
resort_df["Open Pistes km"] = resort_df["Open Pistes km"].str.split("km").str[0]    
```

```python
resort_df["Slope Condition"] = resort_df["Slope Condition"].str.split("Slope condition").str[0]   
```

```python
resort_df["Open Lifts"] = resort_df["Open Lifts"].str.split("Lifts open").str[0]
```

### Numerical information: removing measurement information and renaming column (adding the measuring unit)

```python
resort_df["New snow in resort (past 24h)"] = resort_df["New snow in resort (past 24h)"].str.split("cm").str[0]    
resort_df["New snow in ski area (past 24h)"] = resort_df["New snow in ski area (past 24h)"].str.split("cm").str[0]    

resort_df.rename(columns={"New snow in resort (past 24h)": "New snow in resort (past 24h) - cm"}, inplace=True)
resort_df.rename(columns={"New snow in ski area (past 24h)": "New snow in ski area (past 24h) - cm"}, inplace=True)
```

```python
resort_df["Surface area of pistes"] = resort_df["Surface area of pistes"].str.split("km").str[0]    
resort_df["Pistes covered by lift pass (regional)"] = resort_df["Pistes covered by lift pass (regional)"].str.split("km").str[0]    

resort_df.rename(columns={"Surface area of pistes": "Surface area of pistes km^2"}, inplace=True)
resort_df.rename(columns={"Pistes covered by lift pass (regional)": "Pistes covered by lift pass (regional) - km"}, inplace=True)
```

### Typo: rename columns

```python
for col in resort_df.columns:
    if "resortat" in col:     
        new_col = col.replace("resortat", "resort at")  
        resort_df = resort_df.rename(columns={col: new_col})
```

```python
for col in resort_df.columns:
    if "runsat" in col:
        new_col = col.replace("runsat", "runs at")
        resort_df = resort_df.rename(columns={col: new_col})
```

```python
resort_df.head()
```

### Column merging and dropping: excessive number of similar columns

```python
depth_cols_resort = [col for col in resort_df.columns if "Depth of snow in resort" in col]
depth_cols_runs = [col for col in resort_df.columns if "Depth of snow on upper runs" in col]
```

```python
resort_df["Depth of snow in resort"] = resort_df[depth_cols_resort].apply(lambda x: ",".join(map(str, x.dropna())), axis=1)
resort_df["Depth of snow on upper runs"] = resort_df[depth_cols_runs].apply(lambda x: ",".join(map(str, x.dropna())), axis=1)
```

```python
resort_df.drop(list(resort_df.filter(regex="metres")), axis=1, inplace=True)
```

```python
resort_df["Depth of snow in resort"] = pd.to_numeric(resort_df["Depth of snow in resort"].str.split("cm").str[0], errors="coerce")
resort_df["Depth of snow on upper runs"] = pd.to_numeric(resort_df["Depth of snow on upper runs"].str.split("cm").str[0], errors= "coerce")

resort_df.rename(columns={"Depth of snow in resort": "Depth of snow in resort - cm"}, inplace=True)
resort_df.rename(columns={"Depth of snow on upper runs": "Depth of snow on upper runs - cm"}, inplace=True)

resort_df["Depth of snow in resort - cm"] = resort_df["Depth of snow in resort - cm"].replace(np.nan, "Not Available")
resort_df["Depth of snow on upper runs - cm"] = resort_df["Depth of snow on upper runs - cm"].replace(np.nan, "Not Available")
```

### Date: calculation based on provided time frame

```python
today = datetime.now()

def extract_days(s):
  match = re.search(r"\d+", s)
  return int(match.group()) if match else 0

resort_df["Last snowfall in resort"] = resort_df["Last snowfall in resort"].apply(lambda x: (today - timedelta(days=extract_days(str(x)))).date())
resort_df["Last snowfall in ski area"] = resort_df["Last snowfall in ski area"].apply(lambda x: (today - timedelta(days=extract_days(str(x)))).date())
```

### Date: conversion to datetime format

```python
resort_df["Last snowfall in resort"] = pd.to_datetime(resort_df["Last snowfall in resort"], format = "%d.%m.%Y")
resort_df["Last snowfall in ski area"] = pd.to_datetime(resort_df["Last snowfall in ski area"], format = "%d.%m.%Y")
```

### Metric conversion

```python
def cm_to_inches(cm_value):
   if isinstance(cm_value, float):
       return round(cm_value / 2.54, 2)
   else:
       return cm_value

resort_df["Depth of snow on upper runs - inches"] = resort_df["Depth of snow on upper runs - cm"].apply(cm_to_inches)
resort_df["Depth of snow in resort - inches"] = resort_df["Depth of snow in resort - cm"].apply(cm_to_inches)
```

```python
resort_df.head()
```

### Price categories: merging and dropping

```python
adult_columns = resort_df.filter(like="Adult").columns
young_columns = resort_df.filter(like="Young").columns
senior_columns = resort_df.filter(like="Senior").columns
child_columns = resort_df.filter(like="Child").columns
resort_df[adult_columns]
```

```python
def extract_price(value):
   match = re.search(r"(\d+)\.- CHF", value)
   return np.nan if match is None else float(match.group(1))
```

```python
for column in adult_columns:
    resort_df[column] = resort_df[column].astype(str)
    resort_df[column] = resort_df[column].apply(extract_price)

for column in young_columns:
    resort_df[column] = resort_df[column].astype(str)
    resort_df[column] = resort_df[column].apply(extract_price)

for column in senior_columns:
    resort_df[column] = resort_df[column].astype(str)
    resort_df[column] = resort_df[column].apply(extract_price)

for column in child_columns:
    resort_df[column] = resort_df[column].astype(str)
    resort_df[column] = resort_df[column].apply(extract_price)

resort_df["adult - CHF"] = resort_df[adult_columns].apply(lambda x: ",".join(x.dropna().astype(str)), axis=1)
resort_df["young person - CHF"] = resort_df[young_columns].apply(lambda x: ",".join(x.dropna().astype(str)), axis=1)
resort_df["senior - CHF"] = resort_df[senior_columns].apply(lambda x: ",".join(x.dropna().astype(str)), axis=1)
resort_df["children - CHF"] = resort_df[child_columns].apply(lambda x: ",".join(x.dropna().astype(str)), axis=1)
```

```python
resort_df.head()
```

```python
resort_df["senior - CHF"] = resort_df["senior - CHF"].fillna(resort_df["adult - CHF"])
resort_df["young person - CHF"] = resort_df["young person - CHF"].fillna(resort_df["adult - CHF"])
```

```python
resort_df.drop(list(resort_df.filter(regex = "Adult")), axis = 1, inplace = True)
resort_df.drop(list(resort_df.filter(regex = "Young")), axis = 1, inplace = True)
resort_df.drop(list(resort_df.filter(regex = "Senior")), axis = 1, inplace = True)
resort_df.drop(list(resort_df.filter(regex = "Child")), axis = 1, inplace = True)
```

### Additional check for possible missing values

```python
nan_count_per_column = resort_df.isnull().sum()
nan_count_per_column
```

### Threshold defined to 70 to account for resorts that may not yet be open 

```python
threshold = 70
resort_df = resort_df.dropna(axis=1, thresh = resort_df.shape[0] - threshold)
```

### Missing information from columns relevant to not be dropped, that contain "-" replaced with a note to its unavailability

```python
resort_df.replace("–", "Not Available", inplace=True)
```
